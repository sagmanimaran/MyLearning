```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
#data for x and y axis
x=[1,2,3]
y=[4,1,3]

#plotting the data
plt.plot(x,y)

#naming the chart, legend
plt.title("Demo Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")

#
plt.show()
```

```python
style.use('ggplot')
x1 = [1,2,3]
y1 = [12,16,6]
x2 = [1,2,3]
y2 = [6,15,7]

#plt.plot(data1.data2,'color',label="labelname",linewidth=width)
plt.plot(x1,y1,'g',label='Data 1',linewidth=1)
plt.plot(x2,y2,'b',label='Data 2',linewidth=3)

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

#showing the legend
plt.legend()

#showing the grid
plt.grid(True,color='c')
plt.show()
```

```python
#bar graph
#bar graphs are used to show data on different categories
#while histograms arre used to show the frequency of the data

import numpy as np

x1=np.linspace(1,10,5)
y1=np.linspace(10,100,5)

x2=np.linspace(2,11,5)
y2=np.linspace(20,200,5)

plt.bar(x1,y1,color='g',label="Data 1",width=0.5)
plt.bar(x2,y2,color="b",label="Data2",width=0.5)

plt.legend()

plt.show()
```

```python
#histogram
#histograms are used to represent the frequeny of the data

People_Age=[10,23,53,68,12,35,67,23]
Age_group=[0,10,20,30,40,50,60,70]
plt.hist(People_Age,Age_group,histtype='bar')

plt.plot()
plt.show()
```

```python
#Scatter plot

plt.scatter(x1,y1,color='r',label="Data 1")
plt.scatter(x2,y2,color='g',label="Data 2")

plt.plot()
plt.legend()
plt.show()
```

```python
date=[1,2,3,4,5,6,7]
working_hours=[8,7.30,4,8,8,8.30,9]
lunch_hours=[1,1,1,0.30,0.30,1,1]
tea_hours=[0.15,0.15,0.15,0.15,0.15,0.15,0.10]

plt.plot([],[],color='c',label="Working Hours",linewidth=5)
plt.plot([],[],color='m', label="Lunch Hours",linewidth=5)
plt.plot([],[],color='y', label="Tea Hours",linewidth=5)


plt.stackplot(date, working_hours,lunch_hours,tea_hours, colors=['c','m','y'])

plt.legend()
plt.show()
```
